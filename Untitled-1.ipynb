```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import ttest_ind
import warnings
warnings.filterwarnings('ignore')

np.random.seed(42)
n = 1000

categories = ['Electronics', 'Clothing', 'Books', 'Furniture', 'Sports', 'Beauty']

data = {
    'order_id': range(1, n+1),
    'customer_id': np.random.randint(100, 300, n),
    'category': np.random.choice(categories, n),
    'price': np.round(np.random.uniform(100, 5000, n), 2),
    'discount': np.round(np.random.uniform(0, 0.5, n), 2),
    'delivery_days': np.random.randint(1, 20, n),
    'customer_rating': np.random.randint(1, 6, n),
    'returned': np.random.choice([0, 1], n, p=[0.75, 0.25])
}

df = pd.DataFrame(data)
df.to_csv('ecommerce_data.csv', index=False)

print("Dataset saved!")
print("Shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
print("\nMissing Values:")
print(df.isnull().sum())
df.drop_duplicates(inplace=True)

total_orders = len(df)
total_returned = df['returned'].sum()
return_pct = (total_returned / total_orders) * 100

print(f"\nTotal Orders: {total_orders}")
print(f"Total Returned: {total_returned}")
print(f"Return Percentage: {return_pct:.2f}%")
```

```python
print("=== Price Stats ===")
print(f"Mean:   {df['price'].mean():.2f}")
print(f"Median: {df['price'].median():.2f}")
print(f"Std Dev:{df['price'].std():.2f}")

print("\n=== Delivery Days Stats ===")
print(df['delivery_days'].describe().round(2))

print("\n=== Customer Rating Stats ===")
print(f"Average Rating: {df['customer_rating'].mean():.2f}")

print("\n=== Discount Stats ===")
print(f"Average Discount: {df['discount'].mean():.2f}")
```

```python
# Chart 1: Category vs Return Rate
category_return = df.groupby('category')['returned'].mean() * 100
category_return_sorted = category_return.sort_values(ascending=False)

plt.figure(figsize=(8,5))
category_return_sorted.plot(kind='bar', color='coral', edgecolor='black')
plt.title('Return Rate by Category (%)')
plt.ylabel('Return Rate (%)')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('category_return.png')
plt.show()
print("Chart 1 saved!")

# Chart 2: Delivery Days vs Return (Boxplot)
plt.figure(figsize=(6,5))
sns.boxplot(x='returned', y='delivery_days', data=df, palette='Set2')
plt.title('Delivery Days vs Return')
plt.xticks([0,1], ['Not Returned', 'Returned'])
plt.tight_layout()
plt.savefig('delivery_return.png')
plt.show()
print("Chart 2 saved!")

# Chart 3: Price Histogram
plt.figure(figsize=(7,5))
df['price'].hist(bins=30, color='steelblue', edgecolor='black')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('price_distribution.png')
plt.show()
print("Chart 3 saved!")

# Chart 4: Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df[['price','discount','delivery_days','customer_rating','returned']].corr(),
            annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.savefig('heatmap.png')
plt.show()
print("Chart 4 saved!")
```

```python
# IQR Method
Q1 = df['delivery_days'].quantile(0.25)
Q3 = df['delivery_days'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers_iqr = df[(df['delivery_days'] < lower) | (df['delivery_days'] > upper)]
print(f"Delivery Day Outliers (IQR): {len(outliers_iqr)}")

# Z-Score Method
df['z_price'] = np.abs(stats.zscore(df['price']))
price_outliers = df[df['z_price'] > 3]
print(f"Price Outliers (Z-score): {len(price_outliers)}")

# Boxplot
plt.figure(figsize=(6,4))
sns.boxplot(y=df['delivery_days'], color='lightblue')
plt.title('Delivery Days Outlier Check')
plt.tight_layout()
plt.savefig('outliers_boxplot.png')
plt.show()
print("Outlier chart saved!")
```

```python
returned_days = df[df['returned'] == 1]['delivery_days']
not_returned_days = df[df['returned'] == 0]['delivery_days']

t_stat, p_value = ttest_ind(returned_days, not_returned_days)

print(f"Mean delivery days (Returned):     {returned_days.mean():.2f}")
print(f"Mean delivery days (Not Returned): {not_returned_days.mean():.2f}")
print(f"\nT-Statistic: {t_stat:.4f}")
print(f"P-Value:     {p_value:.4f}")

if p_value < 0.05:
    print("\nResult: Reject H0 — Delivery delay DOES significantly impact returns!")
else:
    print("\nResult: Fail to reject H0 — No significant impact found.")
```

```python
customer_stats = df.groupby('customer_id').agg(
    total_orders=('order_id', 'count'),
    total_returns=('returned', 'sum'),
    avg_discount=('discount', 'mean'),
    avg_rating=('customer_rating', 'mean')
).reset_index()

customer_stats['return_rate'] = customer_stats['total_returns'] / customer_stats['total_orders']

def segment(row):
    if row['return_rate'] > 0.5:
        return 'High-Return Customer'
    elif row['avg_discount'] > 0.35:
        return 'Discount-Driven'
    elif row['avg_rating'] >= 4:
        return 'Premium Customer'
    else:
        return 'Normal Customer'

customer_stats['segment'] = customer_stats.apply(segment, axis=1)

print("Customer Segments:")
print(customer_stats['segment'].value_counts())

plt.figure(figsize=(7,5))
customer_stats['segment'].value_counts().plot(kind='bar', color='purple', edgecolor='black')
plt.title('Customer Segments')
plt.xlabel('Segment')
plt.ylabel('Count')
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig('customer_segments.png')
plt.show()
print("Segment chart saved!")
```
